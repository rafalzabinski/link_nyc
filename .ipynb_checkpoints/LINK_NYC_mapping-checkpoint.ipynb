## LINK_NYC

```python
#Dependencies
import pandas as pd
import csv
import matplotlib.pyplot as plt
from api_keys import g_key
import gmaps
import os
from pathlib import Path
import geopandas as gpd
```

```python
#Importing the NYC Wi-Fi Hostpot Location data
file = Path("raw_data/NYC_Wi-Fi_Hotspot_Locations.csv")
data = pd.read_csv(file)
data.head()
```

```python
#Calculating how many different providers are in NYC
provider = data["Provider"].value_counts(ascending=True)
provider.plot(kind="barh", alpha=0.5,figsize=(8, 8))
plt.title("Hotspots providers")
plt.show()
```

```python
#Calculating how many LinkNYC hostpots in each borough
prov = 'LinkNYC - Citybridge'
cond = (data['Provider'] == prov)
prov_by_borough = data[cond].groupby('Borough Name')["OBJECTID"].count()
prov_by_borough = prov_by_borough.sort_values(ascending=True)
prov_by_borough
```

```python
# Creating the heat map of LinkNYC Hostposts in NYC
prov = 'LinkNYC - Citybridge'
cond = (data['Provider'] == prov)
gmaps.configure(api_key=g_key)
m = gmaps.Map()
locations = data[cond][['Latitude','Longitude']]
weight = data[cond]['OBJECTID']
m = gmaps.figure(map_type = 'HYBRID')
heat = gmaps.heatmap_layer(locations, weights=weight,point_radius=10)
m.add_layer(heat)
m
```

```python
#Calculating how many LinkNYC hostpots in neighborhoods

prov = 'LinkNYC - Citybridge'
cond = (data['Provider'] == prov)
spots_by_neighborhood = data[cond].groupby('Neighborhood Tabulation Area (NTA)')["OBJECTID"].size()
spots_by_neighborhood = spots_by_neighborhood.sort_values(ascending=True)
spots_by_neighborhood.plot(kind="barh", figsize=(15, 45),alpha=0.5, legend = False)
plt.title("HotSpots by neighborhood")
plt.show()
```

```python
#Adding Population Column to the data set
population_file = Path("raw_data/Census_Demographics_at_the_Neighborhood_Tabulation_Area__NTA__level.csv")
pop = pd.read_csv(population_file)
pop.drop(['Geographic Area - Borough', 
         'Geographic Area - 2010 Census FIPS County Code',
         'Total Population 2000 Number',
         'Total Population Change 2000-2010 Number',
         'Total Population Change 2000-2010 Percent'], axis = 1, inplace=True)

pop.rename(columns={'Geographic Area - Neighborhood Tabulation Area (NTA)* Code':
                    'Neighborhood Tabulation Area Code (NTACODE)'},inplace=True)

final_data = pd.merge(pop, data, on = 'Neighborhood Tabulation Area Code (NTACODE)')
final_data.head()
```

```python
#Calculating how many LinkNyc per 10k people in each neighborhood
hotspots = pd.DataFrame({'Neighborhood Tabulation Area (NTA)':spots_by_neighborhood.index, 'No_of_Hotspots':spots_by_neighborhood.values})
hotspots_per_pop = pd.merge(hotspots, final_data, on= 'Neighborhood Tabulation Area (NTA)')
hotspots_per_pop["Hotspots per 10k people"] = (hotspots_per_pop['No_of_Hotspots']/hotspots_per_pop['Total Population 2010 Number'])*10000
hotspots_per_pop = hotspots_per_pop.sort_values(by= ["Hotspots per 10k people"],ascending=False)
hotspots_per_pop.head()
```

```python
#Saving the result in CSV file (just in case)
hotspots_per_pop.to_csv(r'raw_data/Hotspots_per_pop.csv', index = False)
```

```python
#Importing shape file for NYC Neighborhood Tabulation Areas 
sf_path = Path("raw_data/Neighborhood Tabulation Areas (NTA)/geo_export_196ffd4c-8e62-442e-8074-99964842e3a9.shp")
sf = gpd.read_file(sf_path, encoding='utf-8')
nyc_sf = sf
nyc_shape = nyc_sf.to_crs({'init': 'epsg:4326'})
nyc_shape.head()
```

```python
#Plotting the Neighborhoods on NYC map
nyc_shape.plot()
```

```python
#Merging the data for "LinkNyc per 10k people in each neighborhood" MAP
file = Path("raw_data/Hotspots_per_pop.csv")
hotspots_per_pop = pd.read_csv(file)
hotspots_per_pop.rename(columns={'Neighborhood Tabulation Area Code (NTACODE)' : 'ntacode'}, inplace = True)
data = pd.merge(nyc_shape, hotspots_per_pop, on = 'ntacode',how="outer")
data = data.sort_values(by = 'Hotspots per 10k people')
data["Hotspots per 10k people"] = data["Hotspots per 10k people"].fillna(0)
data.head()
```

```python
#"LinkNyc per 10k people in each neighborhood" MAP
ax = data.plot(column='Hotspots per 10k people', cmap='coolwarm', k=5,figsize = (15,10), legend=True)
plt.title("LINK NYC Hotspots per 10k people");
ax.set_axis_off()
plt.savefig('LINKNYC.png')
```
